````python
from pycaret.datasets import get_data
import pandas as pd

df = get_data(dataset='traffic', verbose=False)

df

````

````python
from pycaret.regression import *


# ```
# Explanation:
# data: Your dataset (df).
# target: The target column ('traffic_volume').
# session_id: Random seed for reproducibility.
# use_gpu: Enables GPU if available and compatible.
# exclude: List of model IDs to exclude during the modeling process.

# ```

# Using the check for GPU availability
dataset = setup(
    data=df, 
    target='traffic_volume', 
    session_id=438, 
    use_gpu=True,
)


````

````python
import pandas as pd
from pycaret.regression import models

# Convert models to a DataFrame for better readability
models_df = pd.DataFrame(models())
models_df
````

````python

best = compare_models(exclude=['ard', 'ransac', 'tr', 'kr', 'svm', 'mlp'])
````

````python
best
````

````python
tuned_best = tune_model(estimator=best, 
                        fold=5, 
                        optimize='MAE',
                        n_iter=10)
````

````python
tuned_best = tune_model(estimator = best)
````

````python
final_model = finalize_model(estimator = tuned_best) 
````

````python
save_model(final_model, "./rf_base_traffic")
````
